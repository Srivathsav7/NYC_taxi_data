```python
import pandas as pd
```

```python
df=pd.read_csv('taxi_trip_data.csv')
df=df.iloc[0:7000,]
df.head()
```

```python
df.dropna(inplace=True)
df.drop_duplicates().reset_index(drop=True)
df['pickup_datetime']=pd.to_datetime(df['pickup_datetime'])
df['dropoff_datetime']=pd.to_datetime(df['dropoff_datetime'])
datetime_dim=df[['pickup_datetime','dropoff_datetime']]
datetime_dim.head()
```

```python
datetime_dim.drop_duplicates().reset_index(drop=True)
```

```python
datetime_dim['pick_hour']=datetime_dim['pickup_datetime'].dt.hour
datetime_dim['pick_month']=datetime_dim['pickup_datetime'].dt.month
datetime_dim['pick_year']=datetime_dim['pickup_datetime'].dt.year
datetime_dim['pick_day']=datetime_dim['pickup_datetime'].dt.day
datetime_dim['pick_weekday']=datetime_dim['pickup_datetime'].dt.weekday
datetime_dim['drop_hour']=datetime_dim['dropoff_datetime'].dt.hour
datetime_dim['drop_month']=datetime_dim['dropoff_datetime'].dt.month
datetime_dim['drop_year']=datetime_dim['dropoff_datetime'].dt.year
datetime_dim['drop_day']=datetime_dim['dropoff_datetime'].dt.day
datetime_dim['drop_weekday']=datetime_dim['dropoff_datetime'].dt.weekday
datetime_dim['datetime_id']=datetime_dim.index
datetime_dim=datetime_dim[['datetime_id','pickup_datetime','dropoff_datetime','pick_hour','pick_month','pick_year','pick_day','pick_weekday','drop_hour','drop_month','drop_year','drop_day','drop_weekday']]
datetime_dim.head()
```

```python
passenger_count_dim=df[['passenger_count']]
passenger_count_dim['passenger_count_id']=passenger_count_dim.index
passenger_count_dim=passenger_count_dim[['passenger_count_id','passenger_count']]
passenger_count_dim.head()
```

```python
trip_distance_dim=df[['trip_distance']]
trip_distance_dim['trip_distance_id']=trip_distance_dim.index
trip_distance_dim=trip_distance_dim[['trip_distance_id','trip_distance']]
trip_distance_dim.head()
```

```python
pickup_dim=df[['pickup_location_id']]
pickup_dim['pickup_id']=pickup_dim.index
pickup_dim=pickup_dim[['pickup_id','pickup_location_id']]
pickup_dim.head()
```

```python
dropoff_dim=df[['dropoff_location_id']]
dropoff_dim['dropoff_id']=dropoff_dim.index
dropoff_dim=dropoff_dim[['dropoff_id','dropoff_location_id']]
dropoff_dim.head()
```

```python
d1={
    1:'Standard Rate',
    2:'JFK',
    3:'Newark',
    4:'Nassau',
    5:'Negotiated Fare',
    6:'Group Ride'
}
rate_code_dim=df[['rate_code']]
rate_code_dim['rate_code_id']=rate_code_dim.index
rate_code_dim['rate_code_name']=rate_code_dim['rate_code'].map(d1)
rate_code_dim=rate_code_dim[['rate_code_id','rate_code','rate_code_name']]
rate_code_dim.head()
```

```python
d2={
    1:'Credit Card',
    2:'Cash',
    3:'No charge',
    4:'Dispute',
    5:'Unknown',
    6:'Voided trip'
}
payment_type_dim=df[['payment_type']]
payment_type_dim['payment_type_id']=payment_type_dim.index
payment_type_dim['payment_name']=payment_type_dim['payment_type'].map(d2)
payment_type_dim=payment_type_dim[['payment_type_id','payment_type','payment_name']]
payment_type_dim.head()
```

```python
fact_table=df[['vendor_id','fare_amount','extra','mta_tax','tip_amount','tolls_amount','imp_surcharge','total_amount']]
fact_table['payment_type_id']=payment_type_dim['payment_type_id']
fact_table['rate_code_id']=rate_code_dim['rate_code_id']
fact_table['dropoff_id']=dropoff_dim['dropoff_id']
fact_table['pickup_id']=pickup_dim['pickup_id']
fact_table['trip_distance_id']=trip_distance_dim['trip_distance_id']
fact_table['passenger_count_id']=passenger_count_dim['passenger_count_id']
fact_table['datetime_id']=datetime_dim['datetime_id']
fact_table.head()
```

```python
dfs = {'Sheet1': fact_table, 'Sheet2': payment_type_dim, 'Sheet3': rate_code_dim, 'Sheet4': dropoff_dim, 'Sheet5': pickup_dim, 'Sheet6': trip_distance_dim, 'Sheet7': passenger_count_dim, 'Sheet8': datetime_dim}

# Provide the Excel file name
excel_file_name = 'output.xlsx'

# Create an Excel writer object
with pd.ExcelWriter(excel_file_name) as writer:
    for sheet_name, df in dfs.items():
        df.to_excel(writer, sheet_name=sheet_name, index=False)
```
